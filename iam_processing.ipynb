```python
import os
import numpy as np
import torch
import random
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import cv2
from pathlib import Path
from typing import List, Tuple, Dict

class IAMHandwritingDataset(Dataset):
    """Custom Dataset for IAM Handwriting data"""
    
    def __init__(self, data_dir: str, words_file: str, split: str = 'train', 
                 train_ratio: float = 0.8, val_ratio: float = 0.1, test_ratio: float = 0.1, 
                 img_height: int = 32, transform=None, seed: int = 42,
                 all_samples=None, char_to_idx=None):
        """
        Args:
            data_dir: Root directory of the IAM dataset 
            words_file: Path to the words.txt file
            split: 'train', 'val', or 'test' split
            train_ratio: Ratio of training data (default 0.8)
            val_ratio: Ratio of validation data (default 0.1)
            test_ratio: Ratio of test data (default 0.1)
            img_height: Height to resize images to (width will be proportional)
            transform: Optional transform to apply to images
            seed: Random seed for reproducible splits
            all_samples: Pre-loaded samples (used for consistent splitting)
            char_to_idx: Shared character-to-index mapping
        """
        self.data_dir = Path(data_dir)
        self.words_file = Path(words_file)
        self.split = split
        self.img_height = img_height
        self.transform = transform
        self.seed = seed
        
        # Either use provided samples or parse them
        if all_samples is None:
            # This is the first dataset being created, so parse and shuffle samples
            self.all_samples = self._parse_words_file()
            
            # Apply consistent random shuffling
            random.seed(self.seed)
            random.shuffle(self.all_samples)
        else:
            # Use the pre-shuffled samples passed in
            self.all_samples = all_samples
        
        # Split into train/validation/test sets
        n_samples = len(self.all_samples)
        n_train = int(n_samples * train_ratio)
        n_val = int(n_samples * val_ratio)
        
        if split == 'train':
            self.samples = self.all_samples[:n_train]
        elif split == 'val':
            self.samples = self.all_samples[n_train:n_train+n_val]
        elif split == 'test':
            self.samples = self.all_samples[n_train+n_val:]
        else:
            raise ValueError(f"Invalid split: {split}. Must be 'train', 'val', or 'test'")
        
        # Use provided character mapping or create from scratch
        if char_to_idx is not None:
            self.char_to_idx = char_to_idx
        else:
            # Create character map from ALL samples, not just this split
            self.char_list = self._get_charset(self.all_samples)
            self.char_to_idx = {char: idx+1 for idx, char in enumerate(self.char_list)}
            self.char_to_idx['<BLANK>'] = 0  # Add blank token for CTC loss
        
        print(f"Created {split} dataset with {len(self.samples)} samples")
    
    def _parse_words_file(self) -> List[Dict]:
        """Parse the IAM words.txt file to get image paths and labels"""
        samples = []
        
        with open(self.words_file, 'r') as f:
            for line in f:
                # Skip comments and empty lines
                if line.startswith('#') or not line.strip():
                    continue
                
                parts = line.strip().split(' ')
                if len(parts) >= 9:
                    word_id = parts[0]
                    status = parts[1]
                    
                    # Only include 'ok' status words
                    if status == 'ok':
                        # Create path to image file
                        file_parts = word_id.split('-')
                        form_id = file_parts[0]
                        subform_id = f"{file_parts[0]}-{file_parts[1]}"
                        image_path = self.data_dir / form_id / subform_id / f"{word_id}.png"
                        
                        # Get transcription (last element in the line)
                        transcription = parts[-1]
                        
                        if os.path.exists(image_path):
                            samples.append({
                                'id': word_id,
                                'image_path': str(image_path),
                                'text': transcription
                            })
        
        return samples
    
    def _get_charset(self, samples) -> List[str]:
        """Extract the character set from all text labels"""
        chars = set()
        for sample in samples:
            chars.update(set(sample['text']))
        return sorted(list(chars))
    
    def __len__(self):
        return len(self.samples)
    
    def preprocess_image(self, img: np.ndarray) -> np.ndarray:
        """Preprocess image for model input"""
        # Convert to grayscale if needed
        if len(img.shape) == 3 and img.shape[2] == 3:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Ensure white background and black text
        if np.mean(img) < 127:
            img = 255 - img
        
        # Resize maintaining aspect ratio
        h, w = img.shape
        new_h = self.img_height
        new_w = int(w * (new_h / h))
        img = cv2.resize(img, (new_w, new_h))
        
        # Normalize to [0, 1]
        img = img.astype(np.float32) / 255.0
        
        # Convert to PyTorch tensor format [C, H, W]
        img = torch.FloatTensor(img).unsqueeze(0)
        
        return img
    
    def encode_text(self, text: str) -> Tuple[torch.Tensor, int]:
        """Encode text to sequence of indices"""
        indices = [self.char_to_idx.get(c, 0) for c in text]
        length = len(indices)
        return torch.LongTensor(indices), length
    
    def __getitem__(self, idx):
        sample = self.samples[idx]
        
        # Load image
        img_path = sample['image_path']
        try:
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            
            # Skip invalid images
            if img is None or img.size == 0:
                # Return empty placeholder and flag for filtering
                return None, None, None, True
            
            # Preprocess image
            img_tensor = self.preprocess_image(img)
            
            # Encode text
            text = sample['text']
            encoded_text, text_length = self.encode_text(text)
            
            return img_tensor, encoded_text, text_length, False
            
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
            # Return empty placeholder and flag for filtering
            return None, None, None, True

# Custom collate function to handle variable-length text sequences and skip invalid samples
def collate_fn(batch):
    # Filter out invalid samples
    batch = [sample for sample in batch if not sample[3]]
    
    if not batch:
        return None, None, None
    
    # Unpack batch
    images, texts, lengths, _ = zip(*batch)
    
    # Get max sequence length in this batch
    max_length = max(len(text) for text in texts)
    
    # For images, we need to handle variable widths
    # Determine max width in this batch
    max_width = max(img.shape[2] for img in images)
    
    # Create padded tensors
    batch_size = len(images)
    padded_images = torch.zeros(batch_size, 1, images[0].shape[1], max_width)
    padded_texts = torch.zeros(batch_size, max_length, dtype=torch.long)
    
    # Fill padded tensors
    for i, (img, text) in enumerate(zip(images, texts)):
        c, h, w = img.shape
        padded_images[i, :, :, :w] = img
        padded_texts[i, :len(text)] = text
    
    # Convert lengths to tensor
    lengths = torch.LongTensor(lengths)
    
    return padded_images, padded_texts, lengths

def create_dataloaders(data_dir, words_file, batch_size=32, img_height=32, seed=42):
    """Create training, validation, and test dataloaders with 8:1:1 split"""
    
    # First, create a temporary dataset to parse and shuffle all samples
    temp_dataset = IAMHandwritingDataset(
        data_dir=data_dir,
        words_file=words_file,
        split='train',  # This doesn't matter here
        train_ratio=1.0,  # Not used for splitting yet
        val_ratio=0.0,
        test_ratio=0.0,
        img_height=img_height,
        seed=seed
    )
    
    # Get all parsed and shuffled samples
    all_samples = temp_dataset.all_samples
    
    # Generate a global character map from ALL samples
    char_list = temp_dataset._get_charset(all_samples)
    char_to_idx = {char: idx+1 for idx, char in enumerate(char_list)} 
    char_to_idx['<BLANK>'] = 0  # Add blank token for CTC loss
    
    # Now create real datasets with proper splits using the same samples and char mapping
    train_dataset = IAMHandwritingDataset(
        data_dir=data_dir,
        words_file=words_file,
        split='train',
        train_ratio=0.8, 
        val_ratio=0.1, 
        test_ratio=0.1,
        img_height=img_height,
        seed=seed,
        all_samples=all_samples,
        char_to_idx=char_to_idx  # Share char mapping
    )
    
    val_dataset = IAMHandwritingDataset(
        data_dir=data_dir,
        words_file=words_file,
        split='val',
        train_ratio=0.8, 
        val_ratio=0.1, 
        test_ratio=0.1,
        img_height=img_height,
        seed=seed,
        all_samples=all_samples,  
        char_to_idx=char_to_idx  # Share char mapping
    )
    
    test_dataset = IAMHandwritingDataset(
        data_dir=data_dir,
        words_file=words_file,
        split='test',
        train_ratio=0.8, 
        val_ratio=0.1, 
        test_ratio=0.1,
        img_height=img_height,
        seed=seed,
        all_samples=all_samples,
        char_to_idx=char_to_idx  # Share char mapping
    )

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=0
    )
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        num_workers=0
    )
    
    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        num_workers=0
    )
    
    return train_loader, val_loader, test_loader, train_dataset.char_to_idx

# Usage example
if __name__ == "__main__":
    # Path to your IAM dataset
    data_dir = 'iam_words/iam_words/words'
    words_file = 'iam_words/iam_words/words.txt'
    
    # Create dataloaders with 8:1:1 split
    train_loader, val_loader, test_loader, char_map = create_dataloaders(
        data_dir=data_dir,
        words_file=words_file,
        batch_size=16
    )
    
    # Print dataset information
    print(f"Character map: {char_map}")
    print(f"Vocabulary size: {len(char_map)}")
    
    # Create reverse mapping for decoding
    idx_to_char = {idx: char for char, idx in char_map.items()}
    
    # Test batch loading from each loader
    loaders = {'train': train_loader, 'val': val_loader, 'test': test_loader}
    
    for split_name, loader in loaders.items():
        print(f"\nChecking {split_name} loader:")
        for images, texts, lengths in loader:
            print(f"Batch images shape: {images.shape}")
            print(f"Batch texts shape: {texts.shape}")
            print(f"Batch text lengths: {lengths}")
            
            # Display multiple images from this batch
            num_samples = min(4, images.shape[0])  # Show up to 4 images
            
            plt.figure(figsize=(15, 4 * num_samples))
            
            for i in range(num_samples):
                # Get and display image
                img = images[i, 0].numpy()
                
                # Decode the text
                text_indices = texts[i][:lengths[i]].tolist()
                decoded_text = ''.join([idx_to_char.get(idx, '') for idx in text_indices if idx > 0])
                
                plt.subplot(num_samples, 1, i+1)
                plt.imshow(img, cmap='gray')
                plt.title(f"{split_name.capitalize()} example {i+1} - Text: '{decoded_text}'")
                plt.axis('off')
            
            plt.tight_layout()
            plt.show()
            
            # Only process one batch from each loader
            break
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CNNRNN(nn.Module):
    def __init__(self, num_chars, input_channels=1):
        super(CNNRNN, self).__init__()
        
        # CNN part
        self.cnn = nn.Sequential(
            # Layer 1
            nn.Conv2d(input_channels, 32, kernel_size=5, stride=1, padding=2),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2), stride=(2, 2)),
            
            # Layer 2
            nn.Conv2d(32, 64, kernel_size=5, stride=1, padding=2),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2), stride=(2, 2)),
            
            # Layer 3
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(1, 2), stride=(1, 2)),
            
            # Layer 4
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(1, 2), stride=(1, 2)),
            
            # Layer 5
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(1, 2), stride=(1, 2))
        )
        
        # Calculate CNN output features
        self.lstm_hidden_size = 256
        
        # Fix 1: Add a projection layer to reduce feature dimensionality
        # Instead of directly connecting CNN to RNN, add a 1x1 conv to reduce channels
        self.feature_projection = nn.Conv2d(256, 256, kernel_size=1, stride=1)
        
        # Fix 2: Update LSTM to handle the actual feature size
        # For a 32px height image, after CNN we'll have:
        # Height: 32 -> 16 -> 8 -> 8 -> 8 -> 8 (after all pooling layers)
        # So each time step has 256*8 = 2048 features
        self.rnn_input_size = 256  # We'll fix this in forward() by reshaping properly
        
        # RNN part
        self.lstm = nn.LSTM(
            input_size=self.rnn_input_size,
            hidden_size=self.lstm_hidden_size,
            num_layers=2,
            bidirectional=True,
            batch_first=True
        )
        
        # Final classifier
        self.classifier = nn.Linear(self.lstm_hidden_size * 2, num_chars)
    
    def forward(self, x):
        # CNN feature extraction
        features = self.cnn(x)  # [batch, channels, height, width]
        
        # Get dimensions
        batch_size, channels, height, width = features.size()
        
        # Fix: Apply 1x1 convolution to reduce features
        features = self.feature_projection(features)  # [batch, 256, height, width]
        
        # Fix: Reshape for RNN - properly handle height dimension
        # Option 1: Flatten height into channels
        # features = features.view(batch_size, width, channels * height)
        
        # Option 2: Average across height dimension (recommended)
        features = features.mean(dim=2)  # [batch, channels, width]
        features = features.permute(0, 2, 1)  # [batch, width, channels]
        
        # RNN sequence processing
        rnn_output, _ = self.lstm(features)  # [batch, width, 2*hidden_size]
        
        # Project to character space
        logits = self.classifier(rnn_output)  # [batch, width, num_chars]
        
        # Apply log softmax for CTC loss
        log_probs = F.log_softmax(logits, dim=2)
        
        return log_probs
    
    def predict(self, x):
        log_probs = self.forward(x)
        predictions = torch.argmax(log_probs, dim=2)
        return predictions


# Usage example with CTC loss
def train_batch(model, criterion, optimizer, images, targets, target_lengths):
    """Train the model on a batch of data"""
    optimizer.zero_grad()
    
    # Forward pass
    log_probs = model(images)
    
    # Prepare for CTC loss
    batch_size = log_probs.size(0)
    input_lengths = torch.full(size=(batch_size,), fill_value=log_probs.size(1), dtype=torch.long)
    
    # Compute CTC loss
    loss = criterion(log_probs.transpose(0, 1), targets, input_lengths, target_lengths)
    
    # Backward pass and optimize
    loss.backward()
    optimizer.step()
    
    return loss.item()


# Decoding function for inference
def decode_predictions(predictions, idx_to_char, blank_idx=0):
    """
    Decode the model predictions using CTC best path decoding
    
    Args:
        predictions: Model predictions [batch_size, sequence_length]
        idx_to_char: Mapping from index to character
        blank_idx: Index of the blank token
        
    Returns:
        List of decoded texts
    """
    results = []
    
    for pred in predictions:
        # Remove duplicates
        collapsed = []
        previous = -1
        for p in pred:
            if p != previous:
                collapsed.append(p.item())
            previous = p.item()
        
        # Remove blanks
        decoded = [idx_to_char[idx] for idx in collapsed if idx != blank_idx]
        results.append(''.join(decoded))
    
    return results
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
import numpy as np
import matplotlib.pyplot as plt
import time
from pathlib import Path
import Levenshtein


# Model definition (from previous code)
# ...CNNRNN class definition...

def train_model(
    model, 
    train_loader, 
    val_loader, 
    char_to_idx,
    device,
    num_epochs=50,
    learning_rate=0.001,
    checkpoint_dir='checkpoints',
    early_stopping_patience=10
):
    # Create checkpoint directory
    checkpoint_path = Path(checkpoint_dir)
    checkpoint_path.mkdir(exist_ok=True, parents=True)
    
    # Create index to char mapping for decoding
    idx_to_char = {idx: char for char, idx in char_to_idx.items()}
    blank_idx = 0  # Assumed CTC blank index
    
    # Model to device
    model = model.to(device)
    
    # Initialize optimizer, scheduler and loss function
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)
    criterion = nn.CTCLoss(blank=blank_idx, reduction='mean', zero_infinity=True)
    
    # Training metrics tracking
    train_losses = []
    val_losses = []
    char_error_rates = []
    word_accuracies = []
    
    # Early stopping
    best_val_loss = float('inf')
    best_epoch = 0
    epochs_without_improvement = 0
    
    # Start training
    print(f"Starting training for {num_epochs} epochs...")
    start_time = time.time()
    
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        epoch_train_loss = 0
        train_batches = 0
        
        for images, texts, text_lengths in train_loader:
            # Skip incomplete batches
            if images is None:
                continue
                
            images = images.to(device)
            texts = texts.to(device)
            text_lengths = text_lengths.to(device)
            
            optimizer.zero_grad()
            
            # Forward pass
            log_probs = model(images)
            
            # Prepare CTC loss inputs
            batch_size = log_probs.size(0)
            input_lengths = torch.full(size=(batch_size,), fill_value=log_probs.size(1), 
                                      dtype=torch.long, device=device)
            
            # FIX: Create proper targets tensor by concatenating only actual characters
            targets = []
            for i in range(batch_size):
                length = text_lengths[i]
                targets.append(texts[i, :length])
            targets = torch.cat(targets)
            
            # Compute loss
            loss = criterion(log_probs.transpose(0, 1), targets, input_lengths, text_lengths)
            
            # Backward pass
            loss.backward()
            
            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
            
            # Update weights
            optimizer.step()
            
            epoch_train_loss += loss.item()
            train_batches += 1
            
        # Calculate average training loss
        avg_train_loss = epoch_train_loss / train_batches if train_batches > 0 else float('inf')
        train_losses.append(avg_train_loss)
        
        # Validation phase
        model.eval()
        epoch_val_loss = 0
        val_batches = 0
        all_predictions = []
        all_ground_truths = []
        
        with torch.no_grad():
            for images, texts, text_lengths in val_loader:
                # Skip incomplete batches
                if images is None:
                    continue
                    
                images = images.to(device)
                texts = texts.to(device)
                text_lengths = text_lengths.to(device)
                
                # Forward pass
                log_probs = model(images)
                
                # Prepare CTC loss inputs
                batch_size = log_probs.size(0)
                input_lengths = torch.full(size=(batch_size,), fill_value=log_probs.size(1), 
                                          dtype=torch.long, device=device)
                
                # FIX: Create proper targets tensor for validation too
                targets = []
                for i in range(batch_size):
                    length = text_lengths[i]
                    targets.append(texts[i, :length])
                targets = torch.cat(targets)
                
                # Compute loss
                loss = criterion(log_probs.transpose(0, 1), targets, input_lengths, text_lengths)
                
                epoch_val_loss += loss.item()
                val_batches += 1
                
                # Get predictions for evaluation metrics
                predictions = model.predict(images)
                decoded_preds = decode_predictions(predictions, idx_to_char, blank_idx)
                
                # Extract ground truth texts
                ground_truths = []
                for i, length in enumerate(text_lengths):
                    gt = ''.join([idx_to_char.get(idx.item(), '') for idx in texts[i, :length] if idx > 0])
                    ground_truths.append(gt)
                
                all_predictions.extend(decoded_preds)
                all_ground_truths.extend(ground_truths)
        
        # Calculate validation metrics
        avg_val_loss = epoch_val_loss / val_batches if val_batches > 0 else float('inf')
        val_losses.append(avg_val_loss)
        
        # Calculate Character Error Rate (CER) and Word Accuracy
        cer = calculate_cer(all_predictions, all_ground_truths)
        word_acc = calculate_word_accuracy(all_predictions, all_ground_truths)
        
        char_error_rates.append(cer)
        word_accuracies.append(word_acc)
        
        # Update learning rate based on validation loss
        scheduler.step(avg_val_loss)
        
        # Early stopping check
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_epoch = epoch
            epochs_without_improvement = 0
            
            # Save the best model
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss': avg_val_loss,
                'char_to_idx': char_to_idx
            }, str(checkpoint_path / 'best_model.pth'))
        else:
            epochs_without_improvement += 1
        
        # Print epoch statistics
        print(f"Epoch {epoch+1}/{num_epochs} - "
              f"Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, "
              f"CER: {cer:.4f}, Word Acc: {word_acc:.4f}, "
              f"Time: {(time.time() - start_time) / 60:.2f} min")
        
        # Save regular checkpoint (every 5 epochs)
        if (epoch + 1) % 5 == 0:
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss': avg_val_loss,
                'char_to_idx': char_to_idx
            }, str(checkpoint_path / f'checkpoint_epoch_{epoch+1}.pth'))
        
        # Early stopping
        if epochs_without_improvement >= early_stopping_patience:
            print(f"Early stopping triggered after {epoch+1} epochs. "
                  f"Best performance at epoch {best_epoch+1}.")
            break
    
    # Training complete
    total_time = time.time() - start_time
    print(f"Training completed in {total_time / 60:.2f} minutes.")
    print(f"Best validation loss: {best_val_loss:.4f} at epoch {best_epoch+1}")
    
    # Plot training history
    plot_training_history(train_losses, val_losses, char_error_rates, word_accuracies)
    
    # Load best model for return
    checkpoint = torch.load(str(checkpoint_path / 'best_model.pth'))
    model.load_state_dict(checkpoint['model_state_dict'])
    
    return model, {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'char_error_rates': char_error_rates,
        'word_accuracies': word_accuracies,
        'best_epoch': best_epoch,
        'best_val_loss': best_val_loss
    }

def calculate_cer(predictions, ground_truths):
    """Calculate Character Error Rate using Levenshtein distance"""
    
    total_distance = 0
    total_length = 0
    
    for pred, gt in zip(predictions, ground_truths):
        distance = Levenshtein.distance(pred, gt)
        total_distance += distance
        total_length += len(gt)
    
    if total_length == 0:
        return 0
    
    return total_distance / total_length

def calculate_word_accuracy(predictions, ground_truths):
    """Calculate Word Accuracy (percentage of exactly matching words)"""
    correct = 0
    total = len(ground_truths)
    
    for pred, gt in zip(predictions, ground_truths):
        if pred == gt:
            correct += 1
    
    if total == 0:
        return 0
    
    return correct / total

def plot_training_history(train_losses, val_losses, char_error_rates, word_accuracies):
    """Plot training history metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 1, 1)
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    
    # Plot metrics
    plt.subplot(2, 1, 2)
    plt.plot(char_error_rates, label='Character Error Rate')
    plt.plot(word_accuracies, label='Word Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Rate')
    plt.title('Evaluation Metrics')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

def evaluate_model(model, test_loader, char_to_idx, device):
    """Evaluate model on test set"""
    model.eval()
    
    idx_to_char = {idx: char for char, idx in char_to_idx.items()}
    blank_idx = 0  # Assumed CTC blank index
    
    all_predictions = []
    all_ground_truths = []
    
    with torch.no_grad():
        for images, texts, text_lengths in test_loader:
            # Skip incomplete batches
            if images is None:
                continue
                
            images = images.to(device)
            
            # Get predictions
            predictions = model.predict(images)
            decoded_preds = decode_predictions(predictions, idx_to_char, blank_idx)
            
            # Extract ground truth texts
            ground_truths = []
            for i, length in enumerate(text_lengths):
                gt = ''.join([idx_to_char.get(texts[i, j].item(), '') for j in range(length) if texts[i, j] > 0])
                ground_truths.append(gt)
            
            all_predictions.extend(decoded_preds)
            all_ground_truths.extend(ground_truths)
    
    # Calculate metrics
    cer = calculate_cer(all_predictions, all_ground_truths)
    word_acc = calculate_word_accuracy(all_predictions, all_ground_truths)
    
    print(f"Test results - Character Error Rate: {cer:.4f}, Word Accuracy: {word_acc:.4f}")
    
    # Display some examples
    display_examples(all_predictions, all_ground_truths, min(10, len(all_predictions)))
    
    return {
        'cer': cer,
        'word_accuracy': word_acc,
        'predictions': all_predictions,
        'ground_truths': all_ground_truths
    }

def display_examples(predictions, ground_truths, num_examples):
    """Display prediction examples with ground truth"""
    for i in range(num_examples):
        print(f"Example {i+1}:")
        print(f"Ground truth: '{ground_truths[i]}'")
        print(f"Prediction:   '{predictions[i]}'")
        print(f"Correct:      {predictions[i] == ground_truths[i]}")
        print("--------------------")


# Main execution
if __name__ == '__main__':
    # Setup device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    # Load data
    data_dir = 'iam_words/iam_words/words'
    words_file = 'iam_words/iam_words/words.txt'
    
    train_loader, val_loader, test_loader, char_to_idx = create_dataloaders(
        data_dir=data_dir,
        words_file=words_file,
        batch_size=64,  # Adjust based on your GPU memory
        img_height=32
    )
    
    # Create model
    num_chars = len(char_to_idx)
    model = CNNRNN(num_chars=num_chars, input_channels=1)
    
    # Train model
    trained_model, history = train_model(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        char_to_idx=char_to_idx,
        device=device,
        num_epochs=50,
        learning_rate=0.001,
        checkpoint_dir='checkpoints',
        early_stopping_patience=10
    )
    
    # Evaluate on test set
    test_results = evaluate_model(
        model=trained_model,
        test_loader=test_loader,
        char_to_idx=char_to_idx,
        device=device
    )
```

```python
def visualize_predictions(model, dataloader, char_to_idx, device, num_samples=5):
    """
    Visualize model predictions on sample images
    
    Args:
        model: Trained CNNRNN model
        dataloader: DataLoader containing samples to predict
        char_to_idx: Character to index mapping
        device: Device to run inference on
        num_samples: Number of samples to visualize
    """
    model.eval()  # Set model to evaluation mode
    
    # Create index to char mapping for decoding
    idx_to_char = {idx: char for char, idx in char_to_idx.items()}
    blank_idx = 0  # CTC blank index
    
    # Get a batch of data
    for images, texts, text_lengths in dataloader:
        if images is None:
            continue
            
        # Only take the requested number of samples
        n = min(num_samples, images.size(0))
        images = images[:n].to(device)
        texts = texts[:n]
        text_lengths = text_lengths[:n]
        
        # Make predictions
        with torch.no_grad():
            predictions = model.predict(images)
            decoded_preds = decode_predictions(predictions, idx_to_char, blank_idx)
        
        # Extract ground truth texts
        ground_truths = []
        for i, length in enumerate(text_lengths):
            gt = ''.join([idx_to_char.get(texts[i, j].item(), '') for j in range(length) if texts[i, j] > 0])
            ground_truths.append(gt)
        
        # Calculate character error rate for each sample
        cers = []
        for pred, gt in zip(decoded_preds, ground_truths):
            cer = Levenshtein.distance(pred, gt) / max(len(gt), 1)
            cers.append(cer)
        
        # Display images with predictions
        plt.figure(figsize=(15, 4 * n))
        
        for i in range(n):
            plt.subplot(n, 1, i+1)
            
            # Get and display image (remove channel dimension and transpose if needed)
            img = images[i, 0].cpu().numpy()
            plt.imshow(img, cmap='gray')
            
            # Display prediction and ground truth
            correct = decoded_preds[i] == ground_truths[i]
            color = 'green' if correct else 'red'
            
            title = f"Ground truth: '{ground_truths[i]}'\n"
            title += f"Prediction: '{decoded_preds[i]}'\n"
            title += f"CER: {cers[i]:.2f}"
            
            plt.title(title, color=color)
            plt.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # Only process one batch
        break

# Usage
def check_trained_model(model, test_loader, char_to_idx, device, num_samples=5):
    # Ensure model is loaded properly and on the right device
    model = model.to(device)
    # Visualize predictions
    visualize_predictions(model, test_loader, char_to_idx, device, num_samples)
    
# Example call:
# check_trained_model(trained_model, test_loader, char_to_idx, device, num_samples=8)
```
